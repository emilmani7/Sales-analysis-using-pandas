# Cleaning Data

```python
def nans(df): return df[df.isnull().any(axis=1)]
nans(df)
```

```python
df =df.dropna(how='all')
dff=df.dropna().reset_index(drop=True)
df.head()
```

```python
df['Month'] = df['Order Date'].str[:2]
df.head()  
```

```python
df.head(5)
```

```python
#df['Month'] = df['Month'].astype(int)
```

```python
df = df[df['Month'].str[0:2] != 'Or']
df['Month'] = df['Month'].astype('int32')
df.head()
```

```python
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'], errors='coerce')
df['Price Each'] = pd.to_numeric(df['Price Each'], errors='coerce')
```

```python
df['Sales'] = df['Price Each'] * df['Quantity Ordered']
```

```python
results = df.groupby('Month').sum()
```

So 12 th month had the largest sale

```python
import matplotlib.pyplot as plt
months = range(1,13)
plt.bar(months, results['Sales'])
plt.xticks(months)
plt.xlabel('Months')
plt.ylabel('Sales in USD')
plt.show()
```

```python
#creating a column named city to find out which city had the highest sales
def get_city(address):
    return address.split(',')[1]
def get_state(address):
    return address.split(',')[2].split(' ')[1]
df['City'] = df['Purchase Address'].apply(lambda x:f"{get_city(x)} ({get_state(x)}")
```

```python
sales_city = df.groupby('City').sum()
```

```python
import matplotlib.pyplot as plt

keys = [city for city, df in df.groupby(['City'])]

plt.bar(keys,df.groupby(['City']).sum()['Sales'])
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month number')
plt.xticks(keys, rotation='vertical', size=8)
plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df['Hour'] = df['Order Date'].dt.hour
df['Minute'] = df['Order Date'].dt.minute
df['Count'] = 1
df.head()
```

```python
keys = [pair for pair, df in df.groupby(['Hour'])]

plt.plot(keys, df.groupby(['Hour']).count()['Count'])
plt.xticks(keys)
plt.xlabel('Hour')
plt.ylabel('Number of Orders')
plt.grid()
plt.show()
```

maximize ads during the time between 11 am and 7 pm to increase the likelihood of cutomers buying products.

```python
df1 = df[df['Order ID'].duplicated(keep=False)]
df1['Grouped'] = df1.groupby('Order ID')['Product'].transform(lambda x:','.join(x))
df2 = df1[['Order ID','Grouped']].drop_duplicates()
df2.head()
```

```python
from itertools import combinations
from collections import Counter
count = Counter()

for row in df2['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

for key,value in count.most_common(10):
    print(key, value)
```

Here we find that Iphone and lightening charger are the most commonly purchased together. Which means that if a customer selects the Iphone to checkout then the charger can be sugggested as an add on.

```python
product_group = df.groupby('Product')
quantity_ordered = product_group.sum()['Quantity Ordered']


keys = [pair for pair, df in product_group]
plt.bar(keys, quantity_ordered)
plt.ylabel('Quantity Ordered')
plt.xlabel('Product')
plt.xticks(keys, rotation='vertical', size=8)
plt.show()
```
